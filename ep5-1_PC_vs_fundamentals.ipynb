# ep5-1 exploring correlation b/w price trend, characterized by PC's w/ fundamentals

```python
import pandas as pd
df_SP500 = pd.read_csv('SP500_PCA_strategy gain.csv')
df_SP500.set_index('Symbol',inplace=True)
df_SP500.head(10)
```

I would like to first test the correlation between `GICS Sub-Industry` and `PC1`,`PC2` and `PC3`.  First, is there a way to generate a correlation matrix?  Yes, there is, but perhaps this is not the best way of visualizing the correlation or non-correlation between `GICSSector`, which is a discrete variable and `PC1`, which is a continuous variable.  Let's inspect the statistics of the first principal components of stocks belonging to differnt `GICSSector`.

```python
df_PC_stat_by_sector = df_SP500[['GICSSector','PC1','PC2','PC3']].groupby(by=['GICSSector']).agg(['mean','std','count'])
df_PC_stat_by_sector['PC1'].sort_values(by=['mean'],ascending=False)
```

As can be seen from the table above, the standard deviation of the first principal component of different `GICSSector` do not appear to be significantly different.  However, it does appear that the Health Care and Information Technology are less hit by the COVID pandemic.  Stocks from these sectors were able to sustain continuous growth through the COVID pandemic and therefore, their price trends are characterized by high `PC1` values.  `PC2` and `PC3` were also observed to correlate with `GICSSector` to different extents.

We would also like to observe the correlation between the principal components and `GICS Sub-Industry`.

```python
print(len(set(df_SP500['GICS Sub-Industry'])),'sub-industry categories')
pd.set_option('display.max_rows',10)
df_PC_stat_by_subsector = df_SP500[['GICSSector','GICS Sub-Industry','PC1','PC2','PC3']].groupby(by='GICS Sub-Industry').agg(['mean','std','count'])
df_PC_stat_by_subsector['PC1'].sort_values(by=['mean'],ascending=False).head(10)
```

The challenge of using `GICS Sub-Industry` as an attribute is that there is sometimes not enough population in each category.  Let's hold the thought of using `GICS Sub-Industry` as an attribute.  ***This analysis suggests that `GICSSector` can be a relavent attribute of stocks to include in the predictions of stock price behaviors.***
